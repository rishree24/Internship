### Question 1

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/job-search/")
```

```python
job_title = driver.find_element(By.XPATH,'/html/body/div[1]/header/div[3]/div/div/div[1]/div/input')
job_title.click()
```

```python
designation = driver.find_element(By.XPATH,"/html/body/div/div[3]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
designation.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH,"/html/body/div/div[3]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
search = driver.find_element(By.XPATH,'//div[@class="searchForm_btnWrap_advance__VYBHN"]/button')
search.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
title_tags = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in title_tags[:10]:
    title = i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[:10]:
    location = i.text
    job_location.append(location)
    
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company_tags[:10]:
    company = i.text
    company_name.append(company)
    
experience_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[:10]:
    experience = i.text
    experience_required.append(experience)
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df = pd.DataFrame({'Title':job_title, 'Location':job_location, 'Company Name':company_name, 'Experience':experience_required})
driver.quit()
df
```
